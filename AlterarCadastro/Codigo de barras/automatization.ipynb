```python
import pandas as pd
import pyautogui as py
import pyscreeze as pys
import time

tabela = pd.read_excel('Codigo de barras.xlsx')

tabela_teste = tabela.loc[1162:1189, ['Código', 'conferido', 'CodBarras', 'Local']]

print(len(tabela_teste['Código']))

print(tabela.info())
display(tabela_teste)
```

```python
py.FAILSAFE = True

# clicar no Symplex na barra de tarefa
time.sleep(1)
py.click(x=314, y=756)

fora_prateleira = []

for row in tabela_teste.itertuples(index=False):

    # selecionar caixa de texto com tab para inserir cod interno
    time.sleep(0.5)
    py.press('TAB')

    # inserir cod interno
    py.write(str(row.Código))
    time.sleep(0.5)
    py.press('ENTER')
    time.sleep(1)

    # condição caso o valor de cod de barras seja aceito com 13 digitos
    if len(str(row.CodBarras)) == 13 and str(row.CodBarras).isdigit():
        
        # selecionar a caixa de texto de cod de barras e inseri-lo
        py.click(x=464, y=128)
        py.write(str(row.CodBarras))
        time.sleep(1)
        py.press('ENTER')

        # confirmar alterção
        time.sleep(1)
        confirmarX, confirmarY = pys.locateCenterOnScreen('imagens/confirmar.png')
        py.click(confirmarX, confirmarY)

        # limpar alteraçao
        # time.sleep(1)
        # limparX, limparY = pys.locateCenterOnScreen('imagens/limpar.png')
        # py.click(limparX, limparY)

        # passar segunda caixa de pergunta se houver
        time.sleep(1)
        try:
            pys.locateOnScreen('imagens/sim_nao.png')
            py.press('ENTER')
        except:
            pass

        time.sleep(0.5)
        try:
            pys.locateOnScreen('imagens/sim_nao.png')
            py.press('ENTER')
        except:
            pass
            
    # caso o valor do cod de barras n seja aceito
    else:

    # condição para o que estiver escrito na coluna conferido
        # se for 'fora prateleiar':
        if str(row.conferido) == 'fora prateleira':

            # Adicionar em uma lista
            produto_fora_prateleira = (str(row.Código), str(row.Local))
            fora_prateleira.append(produto_fora_prateleira)

        # se for qualquer outra coisa
        else:

            # selecionar a imagem para gerar cod de barras
            time.sleep(1)
            gerar_barrasX, gerar_barrasY = pys.locateCenterOnScreen('imagens/barras.png')
            py.click(gerar_barrasX, gerar_barrasY)

            # passar primeira caixa de pergunta
            time.sleep(1)
            py.press('RIGHT')
            py.press('ENTER')

            # confirmar alterção
            time.sleep(1)
            confirmarX, confirmarY = pys.locateCenterOnScreen('imagens/confirmar.png')
            py.click(confirmarX, confirmarY)

            # limpar alteraçao
            # time.sleep(1)
            # limparX, limparY = pys.locateCenterOnScreen('imagens/limpar.png')
            # py.click(limparX, limparY)

            # passar segunda caixa de pergunta se houver
            time.sleep(1)
            try:
                pys.locateOnScreen('imagens/sim_nao.png')
                py.press('ENTER')
            except:
                pass

            time.sleep(0.5)
            try:
                pys.locateOnScreen('imagens/sim_nao.png')
                py.press('ENTER')
            except:
                pass
        
```
