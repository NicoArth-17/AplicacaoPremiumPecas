```python
py.FAILSAFE = True

# clicar no Symplex na barra de tarefa
time.sleep(1)
py.click(x=000, y=000)

for row in tabela_teste.itertuples(index=False):

    # selecionar caixa de texto com tab para inserir cod interno
    py.press('TAB')

    # inserir cod interno
    py.write(row.Código)
    time.sleep(1)
    py.press('ENTER')

    # condição caso o valor de cod de barras seja aceito com 13 digitos
    if len(str(row.CodBarras)) == 13 and str(row.CodBarras).isdigit():
        pass
        # selecionar a caixa de texto de cod de barras e inseri-lo
        py.click(x=000, y=000)
        py.write(row.CodBarras)
        time.sleep(1)
        py.press('ENTER')

        # confirmar alterção
        time.sleep(1)
        confirm = py.locateOnScreen('imangens/ CONFIRMAR ', confidence=0.5)
        py.click(py.center(confirm))

        # passar segunda caixa de pergunta se houver
        time.sleep(1)
        erro = py.locateOnScreen('imangens/ ERRO', confidence=0.5)
        if erro:
            py.press('ENTER')
        

    # caso o valor do cod de barras n seja aceito
    else:

    # condição para o que estiver escrito na coluna conferido
        # se for 'fora prateleiar':
        if str(row.conferido) == 'fora prateleira':

            # Adicionar em uma lista

            # passar
            pass

        # se for qualquer outra coisa
        else:

            # selecionar a imagem para gerar cod de barras
            time.sleep(1)
            gerar_barras = py.locateOnScreen('imangens/barras.png', confidence=0.5)
            py.click(py.center(gerar_barras))

            # passar primeira caixa de pergunta
            time.sleep(1)
            py.press('RIGHT')
            py.press('ENTER')

            # confirmar alterção
            time.sleep(1)
            confirm = py.locateOnScreen('imangens/ CONFIRMAR ', confidence=0.5)
            py.click(py.center(confirm))

            # passar segunda caixa de pergunta se houver
            time.sleep(1)
            erro = py.locateOnScreen('imangens/ ERRO', confidence=0.5)
            if erro:
                py.press('ENTER')
        
```
